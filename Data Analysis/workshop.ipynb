```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv("/content/all_data.csv",encoding='iso-8859-1')
data
```

```python
data.dropna(subset='OrderID',inplace=True)
data.head()
```

```python
data.rename(columns={'City.1':'CityCompany','Country.1':'CountryCompany','Phone.1':'PhoneCompany'},inplace=True)
data.head()
```

```python
data['OrderDate']=pd.to_datetime(data['OrderDate'])
data.head()
```

```python
data[['CustomerID','OrderID','SupplierID','OrderNumber']]=data[['CustomerID','OrderID','SupplierID','OrderNumber']].astype(int)
data.head()

#for col in data.columns:
  #if 'ID' in data.columns or 'OrderNumber' in data.columns:
   # data[col]=data[col].astype(int)
#data.head()
```

```python
data.info()
```

```python
data.columns=data.columns.str.strip().str.lower()
data.head()
```

```python
data[['customerid', 'firstname', 'lastname', 'city', 'country', 'orderid', 'ordernumber', 'orderdate', 'totalamount']]
```

```python
sns.histplot(data=data, x='totalamount', kde=True)
plt.xlabel("Total Amount")
plt.ylabel("Frequency")
plt.axvline(data['totalamount'].mean(), color='red', label='Mean')
plt.axvline(data['totalamount'].median(), color='black', label='Median')
plt.legend()
plt.show()
#f=px.histogram(data, x='totalamount',nbins=30,title='Distribution of Total Amount')
#f2=px.pie(data, values='totalamount', names='country', title='Total Amount by Country')
#f.show()
#f2.show()
```

```python
#city_counts = data['city'].value_counts().reset_index()
#city_counts.columns = ['city', 'value']
#px.bar(city_counts, x='city', y='value', title='Distribution of Cities')
px.bar(data_frame=data['city'].value_counts(), title= 'Distribution of cities')
```

```python
total_amount_by_country = data.groupby('country')['totalamount'].sum().reset_index()
px.bar(total_amount_by_country, x= 'country', y= 'totalamount', title= 'Total Amount by Country')

#sns.barplot(data=total_amount_by_country, x='country', y='totalamount', color='skyblue')
#plt.title='Total Amount per country'
#plt.xticks(rotation=45)
#plt.show()
```

```python
total_amount_by_city = data.groupby('city')['totalamount'].sum().reset_index()
px.bar(total_amount_by_city, x= 'city', y= 'totalamount', title= 'Total Amount by City',color='city')
```

```python
fullname=data['firstname']+' '+data['lastname']
data['fullname']=fullname
top_customers = data.groupby('fullname')['totalamount'].sum().reset_index().sort_values(by='totalamount', ascending=False).head(10)
px.bar(top_customers,x='fullname', y='totalamount',color='totalamount',title='Top 10 Customers by Total Amount')
```
