. Assign New Column
Create a new column called Rating_Category:

If Stars >= 4  "Excellent"

else  "Good"



```python
def Rating_Category(columns):
  if columns['Stars']>=4.5:
    return 'Excellent'
  else:
    return 'Low'
df['Rating_Category']=df.apply(Rating_Category,axis=1)
df.head(10)
```
